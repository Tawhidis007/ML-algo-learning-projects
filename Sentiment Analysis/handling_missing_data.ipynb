```python
col_with_missing_value=df.isnull().sum()
```

```python
df["Review Votes"].value_counts()
```

```python
#i wanna know what percentage of the data is null
total_cell=np.product(df.shape)
total_null=col_with_missing_value.sum()
print("percantage of missing value{}".format((total_null/total_cell)*100))
```
